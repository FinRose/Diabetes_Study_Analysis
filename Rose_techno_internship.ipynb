Diabetes is one of the most frequent diseases worldwide and the number of diabetic patients are growing over the years. The main cause of diabetes remains unknown, yet scientists believe that both genetic factors and environmental lifestyle play a major role in diabetes.

A few years ago research was done on a tribe in America which is called the Pima tribe. In this tribe, it was found that the ladies are prone to diabetes very early. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients were females at least 21 years old of Pima tribe.

```python
# To plot a distribution plot
sns.displot(pima['BloodPressure'], kind='kde')
plt.show()

# From my plot i can infer that the blood pressure is within the range of 60 to 80 for a large number of the observations. 
#This means that most people's blood pressure from the plot range from 60 to 80.

# to ignore the warnings 
from warnings import filterwarnings
```

```python
# To calculate the BMI 
pima[pima['Glucose']==pima['Glucose'].max()]['BMI']

# The BMI of the person with the highest glucose is 42.9.
```

```python
# To get the mean, median and mode. 
m1 = pima['BMI'].mean()  # mean
print(m1)
m2 = pima['BMI'].median()  # median
print(m2)
m3 = pima['BMI'].mode()[0]  # mode
print(m3) 

# Here the central measures of tendency are equal.
```

```python
# To get the number of women whose glucose level is above mean value
pima[pima['Glucose']>pima['Glucose'].mean()].shape[0]

# It can be said that the number of women with glucose level above the 32.45 mean is 343. 
```

```python
# To get the number of women with median bloodpressue and BMI below median BMI
pima_diabetes1 = pima[(pima['BloodPressure']==pima['BloodPressure'].median()) & (pima['BMI']<pima['BMI'].median())]
number_of_women=len(pima_diabetes1.axes[0])
print("Number of women:" +str(number_of_women))
```

```python
# To create a pairplot
sns.pairplot(data=pima,vars=['Glucose', 'SkinThickness', 'DiabetesPedigreeFunction'], hue='Outcome')
plt.show()

# pairplot helps to display a set of large data in a set of figure differentiated by axes, the plot shows dependency on only one random variable.
```

```python
# To plot a scatterplot
plt.scatter(x='Glucose',y='Insulin',data=pima)
plt.show()

# From the plot one can say that an increase in glucose causes little changes in insulin levels. There is also an insulin increase as glucose increases at some point.
```

```python
# To plot a boxplot showing the 'Age'
plt.boxplot(pima['Age'])

plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()

# Yes from the plot there are outliers just above the horizontal line.
```

```python
# To plot histogram of women who has diabetes
plt.hist(pima[pima['Outcome']==1]['Age'], bins = 5)
plt.title('Distribution of Age for Women who has Diabetes')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


# From the plot the frequency of the women who have diabetes decreases as the age limit increases and the age limit with the highest incidence of diabetes is between age 22 - 30.
```

```python
# To plot histogram of women who do not have diabetes
plt.hist(pima[pima['Outcome']==0]['Age'], bins = 5)
plt.title('Distribution of Age for Women who do not have Diabetes')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


# From the plot the age limit with least risk of diabetes is between 70 - 80, while age limit 22 - 35 are at high risk of from  suffering diabetes.
```

```python
# To get the interquartile range of the variable
Q1 = pima.quantile(0.25)
Q3 = pima.quantile(0.75)
IQR = Q3 - Q1
print(IQR)

# IQR says the range which most of the data lies.

# It can be calculated by substracting the first quartile from the third quartile.

# Why? It is used in filtering the extreme values that lies far away from the rest of the observation. This plot Visualize same as boxplot.
```

```python
# To find correlation matrix

corr_matrix = pima.iloc[:,0:8].corr()

corr_matrix
```

```python
# To visualize the correlation matrix use heatmap
plt.figure(figsize=(8,8))
sns.heatmap(corr_matrix, annot = True)
plt.show()

# Age and pregnancies are positively correlated.
# Glucose and insulin are positively correlated.
# SkinThickness and BMI are positively correlated.
```
